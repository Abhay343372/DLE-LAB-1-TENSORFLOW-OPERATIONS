```python
import tensorflow as tf
print(tf.__version__)
```

```python
a = tf.constant(5)
print(a)

v=tf.constant([1,2,3,4])
print(v)

m=tf.constant([[1,2],[3,4]])
print(m)

t3 = tf.constant([
    [[1, 2], [3, 4]], [[5, 6], [7, 8]]
])
print(t3)
```

```python
a=tf.constant([10,20,30])
b=tf.constant([50,60,70])
print(a+b)
```

```python
a=tf.constant([5,10,15])
b=tf.constant([6,7,8])
print(a/b)
```

```python
a=tf.constant([5,10,15])
b=tf.constant(6)
print(pow(a,b))
```

```python
# operations on scalor tensors
a=tf.constant(20)
b=tf.constant(5)
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(pow(a,b))
```

```python
#operations on 3D tensors
a = tf.constant([
    [[1, 1], [1, 1]],
    [[2, 2], [2, 2]]
])

b = tf.constant([
    [[5, 5], [5, 5]],
    [[3, 3], [3, 3]]
])

print(a + b)
print(a-b)
print(a*b)
print(a/b)
```
